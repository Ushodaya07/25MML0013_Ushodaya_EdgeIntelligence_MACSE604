```python
import cv2

cap = cv2.VideoCapture("sample_video.mp4")

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("Video",frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Display Video Properties

import cv2

cap = cv2.VideoCapture("sample_video.mp4")

total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

print("Total frames:",total_frames)
print("fps:",fps)
print("width",width)
print("height:",height)

cap.release()
```

# Saving 100th frame

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.rectangle(frame, (100,100),(300,300),(0,255,0),3)
    cv2.circle(frame,(200,200),50,(255,0,0),3)

    cv2.putText(frame,"OpenCV practice",(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)

    cv2.imshow("Shapes",frame)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    edges = cv2.Canny(frame,100,200)

    cv2.imshow("Edges",edges)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    negative = cv2.bitwise_not(frame)

    cv2.imshow("Edges",negative)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Detecting movement
import cv2

cap = cv2.VideoCapture(0)

ret, frame1 = cap.read()
ret, frame2 = cap.read()

while True:
    diff = cv2.absdiff(frame1,frame2)
    gray = cv2.cvtColor(diff,cv2.COLOR_BGR2GRAY)

    blur = cv2.GaussianBlur(gray,(5,5),0)
    _,thresh = cv2.threshold(blur,20,255,cv2.THRESH_BINARY)

    cv2.imshow("Motion",thresh)

    frame1 = frame2
    ret,frame2 = cap.read()
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Extract and Display Frame

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def extract_and_display_single_frame_pixels(video_path):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error: Cannot open video")
        return

    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame from video")
        cap.release()
        return

    cap.release()

    # Convert BGR to RGB for consistent display and processing
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Convert RGB to Grayscale
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    print("\n--- RGB Frame Pixel Values (Top-Left 5x5) ---")
    print(frame_rgb[0:5, 0:5, :])

    print("\n--- Grayscale Frame Pixel Values (Top-Left 5x5) ---")
    print(frame_gray[0:5, 0:5])

    # Display the RGB frame
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(frame_rgb)
    plt.title('First Frame (RGB)')
    plt.axis('off')

    # Display the Grayscale frame
    plt.subplot(1, 2, 2)
    plt.imshow(frame_gray, cmap='gray')
    plt.title('First Frame (Grayscale)')
    plt.axis('off')
    plt.show()

# Example usage with your video path
video_path = "Edge-Sample-Video.mp4"
extract_and_display_single_frame_pixels(video_path)
```
